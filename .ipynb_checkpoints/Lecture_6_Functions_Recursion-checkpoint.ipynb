```python
##Function Types(Built in Function and User Define Function)
# Wrute a Function to print the lenght of a list  

cities = ["Dhaka", "London", "Cumilla", "Khulna", "Rangpur", "Rajshahi"]
heroes = ["Thor", "SRK", "Allu_Arjun", "RDJ", "Krish"]

def print_len(list):
    print(len(list))

print_len(cities)
print_len(heroes) 

```

```python
# Write a Function to print the elements of a list in a single line

cities = ["Dhaka","Rajshahi","Khulna","Rangpur","Comilla",]
heroes  = ["Hulk","Iron_Man","Thor","Spider_Man","Krrish","Jervis"]

print(cities[0],end=" ")
print(heroes[1],end=" ")

def print_len(list):
    print(len(list))

def print_list(list):
    for item in list:
        print (item, end= " ")

print_list(heroes)
    
```

```python
# Write a Function to find the factorial of n (n is the parameter)

def cal_fact(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i 
    print (fact)

cal_fact (5)        
        
```

```python
# Write a Function to convert USD to Taka

def converter(usd_val) :
    bdt_val = usd_val * 123
    print(usd_val,"USD =", bdt_val, "BDT")

converter(5)
    
```

```python
### Recursion,,, when a function calls its self repeatedly called recursion.

def show(n):
    if (n==0):
        return
    print(n)
    show(n-1)
show(5)
```

```python
##Recursion vis afactorial. 5! =1*2*3*4*5

def fact(n):
    if (n==1 or n== 0):
        return 1
    return fact(n-1)*n

print(fact(4))
    
```

```python
#WAP a recursive function to calculate the sum of first n natural number (I write a programe that calculate 1 to 5 total number sum, like 1+2+3+4+5=15)

def cal_sum (n):
    if (n == 0 ):
        return 0
    return cal_sum(n-1) + n

sum = cal_sum(5)
print(sum)
```

```python
# Write a recursive function to print all the elements in a the list. 

def print_list(list, index):
    if (index == len(list)):
        return
    print(list[index])
    print_list(list, index+1)

fruits = ["Mango","Orange","Apple","Banana","Palm"]    
print_list(fruits)
```
